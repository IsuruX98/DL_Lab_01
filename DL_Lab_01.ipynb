# **Task 01**

Random 4x4 array from an exponential distribution

```python
import numpy as np

# Create a random 4x4 array from an exponential distribution with lambda=1.0
array_4x4 = np.random.exponential(scale=1.0, size=(4, 4))
print("4x4 Array from Exponential Distribution:\n", array_4x4)
```

Random 100000x1 array and visualize histograms

```python
import matplotlib.pyplot as plt

# Create a random 100000x1 array from an exponential distribution
exp_data = np.random.exponential(scale=1.0, size=100000)

# Create uniform and normal distribution data
uniform_data = np.random.uniform(low=-2, high=2, size=100000)
normal_data = np.random.normal(loc=0, scale=1, size=100000)

# Plot histograms for each distribution
plt.figure(figsize=(12, 6))
plt.hist(exp_data, bins=100, density=True, alpha=0.5, label='Exponential', color='red')
plt.hist(uniform_data, bins=100, density=True, alpha=0.5, label='Uniform', color='green')
plt.hist(normal_data, bins=100, density=True, alpha=0.5, label='Normal', color='blue')

plt.axis([-2.5, 6, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

3D Plot of Z = X² + Y² using Matplotlib

```python
from mpl_toolkits.mplot3d import Axes3D

# Define x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = x**2 + y**2

# Plotting
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

ax.set_title('3D Plot of Z = X² + Y²')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

Calculate Pearson and Spearman Correlations and Visualize

```python
import pandas as pd
import seaborn as sns

# Sample data for demonstration, replace it with the actual data
data = np.random.rand(100, 6)  # Replace with actual data
columns = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
df = pd.DataFrame(data, columns=columns)

# Calculate Pearson and Spearman correlations
pearson_corr = df.corr(method='pearson')
spearman_corr = df.corr(method='spearman')

# Visualize with heatmaps
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', cbar_kws={'label': 'Correlation'})
plt.title('Pearson Correlation Heatmap')

plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', cbar_kws={'label': 'Correlation'})
plt.title('Spearman Correlation Heatmap')

plt.tight_layout()
plt.show()
```

# **Task 02**

Download the 'Au_nanoParticle_dataset.csv' and load the data as a DataFrame

```python
import pandas as pd

# Load the dataset
url = "/content/Au_nanoparticle_dataset.csv"
df = pd.read_csv(url)

# Display the dataframe's first few rows to verify loading
print(df.head())
```

New dataframe by filtering specific columns

```python
# Filter the dataframe for specific columns
filtered_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

Display the first 20 samples

```python
# Display the first 20 samples
print(filtered_df.head(20))
```

Calculate the mean, standard deviation, and quartile values

```python
# Calculate the mean, standard deviation, and quartiles
mean_values = filtered_df.mean()
std_values = filtered_df.std()
quartile_values = filtered_df.quantile([0.25, 0.5, 0.75])

print("Mean:\n", mean_values)
print("Standard Deviation:\n", std_values)
print("Quartiles:\n", quartile_values)
```

Plot histograms for each feature in a 1x4 layout

```python
# Plot histograms for each feature
plt.figure(figsize=(16, 4))

for i, column in enumerate(filtered_df.columns, 1):
    plt.subplot(1, 4, i)
    plt.hist(filtered_df[column], bins=30, color='skyblue', edgecolor='black')
    plt.title(f'{column} Histogram')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Visualize scatter plots and histograms using Seaborn's PairGrid

```python
# Visualize using PairGrid
g = sns.pairplot(filtered_df)
plt.show()
```

Customize PairGrid with histograms and KDE plots

```python
# Customized PairGrid
g = sns.PairGrid(filtered_df)
g.map_upper(sns.histplot)  # Bivariate histogram
g.map_diag(sns.histplot, kde=True)  # Histogram with KDE on diagonals
g.map_lower(sns.kdeplot)  # Bivariate KDE plot

plt.show()
```
